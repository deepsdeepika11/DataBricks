```python
#Transformation :filter, group by, reduce by, select, average,join
# action count ,collect ,display,show

from pyspark.sql.functions import *;
from pyspark.sql.types import *;
data=[
("Joshi",26,'Delhi'),
("Sameer",28,'Karnataka'),
("Avinash",32,"Bihar")

]

schema = StructType([

    StructField("Name",StringType(),True),
    StructField("Age",IntegerType(),True),
    StructField("State",StringType(),True)
])

df = spark.createDataFrame(data,schema)
```

```python
df = df.filter(col('Age')>30)
```

```python
df = df.select(col('Name'))
df.display();
```

```python
df.explain()
```

```python
from pyspark.sql.functions import *;
from pyspark.sql.types import *;
data=[
("Joshi",26,'Delhi'),
("Sameer",28,'Karnataka'),
("Avinash",32,"Bihar"),
("Sam",26,'Telangana'),
("Mohit",28,'Karnataka'),
("Rohit",32,"Bihar")

]

schema = StructType([

    StructField("Name",StringType(),True),
    StructField("Age",IntegerType(),True),
    StructField("State",StringType(),True)
])

df = spark.createDataFrame(data,schema)
```

```python
df.createOrReplaceTempView('Students')
df.display();
```

```python
%sql
Select * from Students
```

```python
%sql
SELECT State, COUNT(*)
FROM Students
GROUP BY State
ORDER BY 1 desc
```

```python
df.groupBy(col('State')).count().orderBy(col('count')).display()

df.display()
```
